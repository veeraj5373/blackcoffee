```python
import pandas as pd
import requests
import bs4 as BeautifulSoup
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import string
# from textblob import TextBlob
import csv
import numpy as np
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
dataframe = pd.read_excel('Input.xlsx', engine='openpyxl', index_col=0)
dataframe
```

```python
li = [url for url in dataframe['URL']]
li
```

```python
text = []
for url in li:
  text.append(requests.get(url,headers={"User-Agent": "XY"}))
```

```python


from bs4 import BeautifulSoup
articles = []
for i in range(len(text)):
    if text[i].content is not None:
        soup = BeautifulSoup(text[i].content, 'html.parser')

        # Find the element with class 'td-post-content'
        post_content = soup.find(class_='td-post-content')

        # Extract text from the found element
        if post_content:
            article_text = post_content.get_text()
            articles.append(article_text)
        else:
            articles.append('no content')
            

            
        
```

```python
len(articles)
```

```python
for i in range(len(articles)):
  articles[i]= articles[i].replace('\n','')
```

```python
stop_words = list(set(stopwords.words('english')))
```

```python
sentences = []
for article in articles:
  sentences.append(len(sent_tokenize(article)))
```

```python
cleaned_articles = [' ']*len(articles)
```

```python
for i in range(len(articles)):
  for w in stop_words:
    cleaned_articles[i]= articles[i].replace(' '+w+' ',' ').replace('?','').replace('.','').replace(',','').replace('!','')
    #? ! , .
```

```python
words = []
for article in articles:
  words.append(len(word_tokenize(article)))
```

```python
words_cleaned = []
for article in cleaned_articles:
  words_cleaned.append(len(word_tokenize(article)))
```

```python

positive_words = pd.read_csv('positive-words.txt', header=None, names=['words'], delimiter='\t', encoding='latin-1')
positive_words
```

```python
negative_words= pd.read_csv('negative-words.txt', header=None, names=['words'], delimiter='\t', encoding='latin-1')

negative_words
```

```python
cleaned_articles[5]
```

```python
positive_score = [0]*len(articles)
for i in range(len(articles)):
  for word in positive_words['words']:
    for letter in cleaned_articles[i].lower().split(' '):
      if letter==word:
        positive_score[i]+=1
```

```python
negative_score
```

```python
negative_score = [0]*len(articles)
for i in range(len(articles)):
  for word in negative_words['words']:
    for letter in cleaned_articles[i].lower().split(' '):
      if letter==word:
        negative_score[i]+=1
```

```python
words_cleaned = np.array(words_cleaned)
sentences = np.array(sentences)
```

```python

dataframe['POSITIVE SCORE'] = positive_score
dataframe['NEGATIVE SCORE'] = negative_score
```

```python
dataframe['POLARITY SCORE'] = (dataframe['POSITIVE SCORE']-dataframe['NEGATIVE SCORE'])/ ((dataframe['POSITIVE SCORE'] +dataframe['NEGATIVE SCORE']) + 0.000001)
dataframe['SUBJECTIVITY SCORE'] = (dataframe['POSITIVE SCORE'] + dataframe['NEGATIVE SCORE'])/( (words_cleaned) + 0.000001)
dataframe['AVG SENTENCE LENGTH'] = np.array(words)/np.array(sentences)
```

```python
complex_words = []
sylabble_counts = []
```

```python
for article in articles:
  sylabble_count=0
  d=article.split()
  ans=0
  for word in d:
    count=0
    for i in range(len(word)):
      if(word[i]=='a' or word[i]=='e' or word[i] =='i' or word[i] == 'o' or word[i] == 'u'):
           count+=1
#            print(words[i])
      if(i==len(word)-2 and (word[i]=='e' and word[i+1]=='d')):
        count-=1;
      if(i==len(word)-2 and (word[i]=='e' and word[i]=='s')):
        count-=1;
    sylabble_count+=count    
    if(count>2):
        ans+=1
  sylabble_counts.append(sylabble_count)
  complex_words.append(ans) 
```

```python
dataframe['PERCENTAGE OF COMPLEX WORDS'] = np.array(complex_words)/np.array(words)
dataframe['FOG INDEX'] = 0.4 * (dataframe['AVG SENTENCE LENGTH'] + dataframe['PERCENTAGE OF COMPLEX WORDS'])
dataframe['AVG NUMBER OF WORDS PER SENTENCES'] = dataframe['AVG SENTENCE LENGTH']
dataframe['COMPLEX WORD COUNT'] = complex_words
dataframe['WORD COUNT'] = words
dataframe['SYLLABLE PER WORD'] = np.array(sylabble_counts)/np.array(words)
```

```python
total_characters = []
for article in articles:
  characters = 0
  for word in article.split():
    characters+=len(word)
  total_characters.append(characters)  
```

```python
personal_nouns = []
personal_noun =['I', 'we','my', 'ours','and' 'us','My','We','Ours','Us','And'] 
for article in articles:
  ans=0
  for word in article:
    if word in personal_noun:
      ans+=1
  personal_nouns.append(ans)
```

```python
dataframe['PERSONAL PRONOUN'] = personal_nouns
dataframe['AVG WORD LENGTH']= np.array(total_characters)/np.array(words)
```
